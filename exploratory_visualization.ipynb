## Define function

The function below must be used to plots the stats.

```python
def plot_stats(fg, tp, ft, reb, ast, to, stl, blk):
    import matplotlib.pyplot as plt
    import numpy as np
    
    team = ('H', 'A')
    y_pos = np.arange(len(team))
    
    plt.clf()
    
    plt.figure(figsize=(14, 6))

    plt.subplot(2, 4, 1)
    plt.title('I')
    plt.bar(y_pos, fg, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(36, 37.2)

    plt.subplot(2, 4, 2)
    plt.title('II')
    plt.bar(y_pos, tp, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(6, 7)

    plt.subplot(2, 4, 3)
    plt.title('III')
    plt.bar(y_pos, ft, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(18, 19)

    plt.subplot(2, 4, 4)
    plt.title('IV')
    plt.bar(y_pos, reb, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(41, 43)

    plt.subplot(2, 4, 5)
    plt.title('V')
    plt.bar(y_pos, ast, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(21, 22)

    plt.subplot(2, 4, 6)
    plt.title('VI')
    plt.bar(y_pos, to, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(13, 14)

    plt.subplot(2, 4, 7)
    plt.title('VII')
    plt.bar(y_pos, stl, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(7, 8)

    plt.subplot(2, 4, 8)
    plt.title('VIII')
    plt.bar(y_pos, blk, align='center', alpha=0.5)
    plt.xticks(y_pos, team)
    plt.ylim(4, 5)

    plt.show()
```

## Data from the last game

When the home team won:

```python
fg = [36.98, 36.6]
tp = [6.7, 6.46]
ft = [18.48, 18.39]
reb = [42.09, 41.92]
ast = [21.65, 21.25]
to = [13.61, 13.83]
stl = [7.5, 7.51]
blk = [4.9, 4.81]

plot_stats(fg, tp, ft, reb, ast, to, stl, blk)
```

```python
fg = [36.47, 37.04]
tp = [6.4, 6.78]
ft = [18.21, 18.53]
reb = [41.79, 42.18]
ast = [21.03, 21.72]
to = [13.87, 13.7]
stl = [7.59, 7.4]
blk = [4.79, 4.92]

plot_stats(fg, tp, ft, reb, ast, to, stl, blk)
```
