## Away victories

As pointed in the report document, the final model tends to predict home team victories and have difficult to predict correctly an away team victory. In the following code cell, I will plot a bar chart to better visualize this.

```python
import matplotlib.pyplot as plt
import numpy as np

datasets = ('Last game', 'Last two games', 'Last five games')
y_pos = np.arange(len(datasets))
correct_away_victories = [49/(431 + 49) * 100, 61/(419 + 61) * 100, 70/(410 + 70) * 100]
correct_home_victories = [636/(35 + 636) * 100, 618/(53 + 618) * 100, 609/(62 + 609) * 100]

plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.bar(y_pos, correct_home_victories, align='center', alpha=0.5)
plt.xticks(y_pos, datasets)
plt.ylabel('Percentage')
plt.title('Correctly predicted home victories')

plt.subplot(1, 2, 2)
plt.bar(y_pos, correct_away_victories, align='center', alpha=0.5)
plt.xticks(y_pos, datasets)
plt.ylabel('Percentage')
plt.title('Correctly predicted away victories')

plt.show()
```

## Classifiers accuracy

As pointed in the report document, MLP classifier had the best accuracies in all the datasets. Below, I will plot the classifiers accuracies in all datasets.

```python
classifiers = ('LR', 'DT', 'KNN', 'GNB', 'MLP')
y_pos = np.arange(len(classifiers))
lg_accs = [61.2, 61.2, 59.1, 61.3, 61.6]
l2g_accs = [61.3, 61.3, 58.2, 61.5, 61.9]
l5g_accs = [61.4, 61.5, 59.8, 61.5, 61.7]

plt.figure(figsize=(10, 3))

plt.subplot(1, 3, 1)
plt.bar(y_pos, lg_accs, align='center', alpha=0.5)
plt.xticks(y_pos, classifiers)
plt.ylabel('Accuracy')
plt.ylim(57.5, 62.5)
plt.title('Last game dataset')

plt.subplot(1, 3, 2)
plt.bar(y_pos, l2g_accs, align='center', alpha=0.5)
plt.xticks(y_pos, classifiers)
plt.ylim(57.5, 62.5)
plt.title('Last two games dataset')

plt.subplot(1, 3, 3)
plt.bar(y_pos, l5g_accs, align='center', alpha=0.5)
plt.xticks(y_pos, classifiers)
plt.ylim(57.5, 62.5)
plt.title('Last five games dataset')

plt.show()
```
